## Setting Up SQL

This is documentation on how to install SQL SSMS (SQL Service Management Studio). Will provide steps on how to provide a mock data to the database. Also, will provide steps on how to authenticate with SQL Server Authentication mode - utilizing a special username for the database.

Steps to install SQL Server (SSMS):
* Install at https://learn.microsoft.com/en-us/ssms/install/install - this will only install the studio manager.
* Install the SQL Server Engine (SQL Server Developer) - this is needed to run SQL SSMS
* Install both execution files, there is no .msi file to run directly.
* Open SSMS - it'll ask you to connect to a dataabse. Perform the following actions...
    * Server Name field - put in "localhost" - this is the same as 127.0.0.1
    * Authentication field - Leave it as Window Authentication (This will change later0)
    *  User Name field - should be greyed out and good as is.
    * Password field - NA
    * Database Name field - leave as default
    * Encrypt field - leave as mandatroy, tick the "Trust Server Certificate" check box.

Importing Mock Data:
* Find a mock dataset - importing a .bak (backup) file is what we will be using.
* download the bak file into your file explorer.
* Follow the path to find your instance of SQL Server.
C:\Program Files\Microsoft SQL Server\MSSQL17.MSSQLSERVER\MSSQL
* Within the file path, there should be a "Backup" file. Cut you bak file you downloaded, and paste it into the "Backup" directory.
* Go back to SQL SSMS - under your server, right-click "Database" and select "Restore Database..."
* Under the Source field, select "Device" > click on the elipses (...).
* On the Select backup devices page click "Add" > select the bak file  on the right pane > click "Ok".

Docker & Metabase Setup:
* Install Docker Desktop.
    * Also a good idea to install Unbutu - we'll utilize the unix terminal to run most of our docker images.
* We can run a Metabase image with the following scripts.
docker pull metabase/metabase:latest  # To run the latest image
docker run -d -p 3000:3000 --name Metabase metabase/metabase  # Run the image in a container.
* In Docker - on the left-hand pane, select Container > Access the web portal via click on the port number "3000:3000"
