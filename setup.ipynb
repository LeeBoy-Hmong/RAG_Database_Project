## Setting Up SQL

This is documentation on how to install SQL SSMS (SQL Service Management Studio). Will provide steps on how to provide a mock data to the database. Also, will provide steps on how to authenticate with SQL Server Authentication mode - utilizing a special username for the database.

Steps to install:
* Install at https://learn.microsoft.com/en-us/ssms/install/install - this will only install the studio manager.
* Install the SQL Server Engine (SQL Server Developer) - this is needed to run SQL SSMS
* Install both execution files, there is no .msi file to run directly.
